(a) Read the text file “speeches.txt” and find the list of unique words with their frequency of occurrence.

# Room Cleaning Robot (Dynamic Control Task)

Consider a Room Cleaner Robot which cleans a room containing dirt. A room can be treated as a $m \times n$ grid with walls on all sides. 

(a) Implement an environment which takes as inputs the values $m$ and $n$. Initially, 10 random cells in the grid contain dirt. The list $d_t$ with dimensions $m \times n$ contains the information on the amount of dirt in each cell of the grid. At any point of time, a unit dirt is added at a random location, i.e., a location $(x, y)$ is picked uniformly at random and $d_t(x, y)$ is updated to $d_t(x, y) + 1$.

```python
import random
m = int(input("Enter m : "))    #taking the input of the size of the environment matrix
n = int(input("Enter n : "))
dt = [[0]*n for i in range(m)]    #declaring the empty matrix to save the data of dirt throughout the environment

i = 0
while(1):
    a = random.choice(range(m))    #randomly choosing the coordinates to add dirt
    b = random.choice(range(n))
    if(dt[a][b] == 0):    #if there is no dirt, add 1 unit of dirt
        dt[a][b] = 1
        i += 1
    if(i == 10):    #the loop stops at i=10, i.e. the dirt is present at 10 places
        break

a = random.choice(range(m))    #randomly choosing the coordinates for the new dirt
b = random.choice(range(n))
dt[a][b] += 1   #incrementing the dirt present by 1

print(dt)    #printing the resulting matrix
```

(b) Implement an agent which at each time instant $t$ observes its position $o_t =(x_t, y_t)$  (but not the dirt information in this location) and performs a random action $a_t \in \{up,down,right,left, pickDirt\}$. On hitting a wall or picking dirt from the current location $(x_t, y_t)$, the agent stays in the same position $(x_t, y_t)$, otherwise, its position changes according to the action. Every action is associated with a reward (or penalty) defined as follows.

   
$$r_t = R((x_t, y_t), a_t) = 
     \begin{cases}
       -1, &\text{if the agent tries to pick dirt and } (x_t, y_t) \text{ is a clean cell}\\
      -10, &\text{if the agent hits a wall}
       \\ 
       d, &\text{if } (x_t, y_t) \text{ has } d \text{ units of dirt and the agent picks it}\\
       0, &\text{otherwise }
     \end{cases}
$$



```python
import random
def action(r, c, ch, score, m, n, dt):
    print("The coordinates of the robot are " + str(r) + " " + str(c))
    if(ch == 0):    #if the random action is UP
        print("The action is UP")
        r -= 1
        if(r < 0):    #if the agent hits the wall
            score -= 10
            r = 0
    elif(ch == 1):    #if the random action is DOWN
        print("The action is DOWN")
        r += 1
        if(r > m-1):    #if the agent hits the wall
            score -= 10
            r = m-1
    elif(ch == 2):    #if the random action is RIGHT
        print("The action is RIGHT")
        c += 1
        if(c > n-1):    #if the agent hits the wall
            score -= 10
            c = n-1
    elif(ch == 3):    #if the random action is LEFT
        print("The action is LEFT")
        c -= 1
        if(c < 0):    #if the agent hits the wall
            score -= 10
            c = 0
    else:    #if the random action is PICK UP DIRT
        print("The action is PICK UP DIRT")
        if(dt[r][c] > 0):    #if the dirt is present at that location
            score += dt[r][c]     #score increases
            dt[r][c] = 0    #and dirt gets reset
        else:
            score -= 1    #else score gets reduced by 1
    return [r, c, score]    #returning the new coordinates and score

score = 0    #initialising the value of the score at 0

r = random.choice(range(m))    #choosing random initial coordinates of the robot
c = random.choice(range(n))
ch = random.choice(range(5))    #choosing random action, 
new = action(r, c, ch, score, m, n, dt)    #storing the returned list into a temporary list

r = new[0]    #updating to new coordinates
c = new[1]
score = new[2]    #updating the score

print("The score after 1 action is " + str(score))    #printing the score after 1 action
print(dt)    #printing the final matrix
```

# Cricket Match Simulation (Stochastic Control Task)

Consider the following simplified variant of the first innings of a one-day cricket match. The innings consists of 300 balls and at any point of time, treat the pair of batsmen playing as a single player. That is, there are 10 players: $(1, 2),(2, 3), \ldots ,(10, 11)$ with exactly one of them batting. In each ball, there are 5 possible shots $A=\{1, 2, 3, 4, 6\}$ and  each of these shots is associated with a risk of the player getting out. This risk varies from player to player. The probabilities of getting out for Player $(1, 2)$ is $pOutMin = [0.01, 0.02, 0.03, 0.1, 0.3]$ and for Player $(10, 11)$ is $pOutMax = [0.1, 0.2, 0.3, 0.5, 0.7]$,  where the $i$th entry is for the $i$th action. If there are $w$ wickets in hand, then use the formula $pOut(a, w) = pOutMax(a) - (pOutMax(a)-pOutMin(a)) \times (w- 1)/9)$ to calculate the probabilities of getting out for Player $(11-w, 11-w+1)$ where $w \in [10]$. Note that $pOut(a, 1)=pOutMin(a)$ and $pOut(a, 10)=pOutMax(a)$. Taking $pRunMin = 0.5$, and $pRunMax = 0.8$,  when the player is not getting out, the probability of successfully obtaining the runs for that shot is given by $pRun(w) = pRunMin + (pRunMax - pRunMin) \times ((w - 1)/9).$



(a) Implement an environment that maintains the state $s_t = (b_t, w_t)$ where $b_t$ is the balls left and $w_t$ is the wickets left at time $t$. Initialize the start state to $s_1 = (300, 10)$. Write a function that accepts input as $a_t \in A$ and returns $r_t$ (the runs scored on that shot) and $s_t$, and also updates $s_{t+1}$. 

```python
import random
def action(a, s, w):    #defining the action function to choose a random action
    
    Prun = []    #this list will store all the probabilities of the runs scored
    for i in range(5):
        Prun.append(0.5+(0.3*((w-1)/9)))    #using the given relation to find the probabilities
    
    Pmin = [0.01, 0.02, 0.03, 0.1, 0.3]    #saving the Pmin and Pmax
    Pmax = [0.1, 0.2, 0.3, 0.5, 0.7]
    Pout = Pmax[a] - (Pmax[a] - Pmin[a])*((w-1)/9)    #Pout as per the given formula
    
    out = random.choices([0, 1], weights=(Pout, 1 - Pout), k=1)    #randomly choosing if the player is out or not as per the given probability
    ap = random.choices([0, 1], weights=(Prun[a], 1 - Prun[a]), k=1)    #randomly choosing if the player has scored or not as per the given probability

    if(out[0] == 0):    #if the player is out
        w -= 1    #decrease the number of wickets by 1
    else:
        if(ap[0] == 0):    #if the player has scored
            s += runs[a]    #update the score

    return [s, w]    #returning the new runs and wickets as per the outcome

s = 0    #initialising the score to the value 0
w = 10    #giving the initial number of wickets as 10
runs = [1, 2, 3, 4, 6]    #this list will help store the possible values of runs scored in each ball

for i in range(6):    #playing 300 balls of the match
    a = int(input("Enter action : "))    #taking the action as input
    if a not in runs:
        print("Invalid action")
        continue
    ac = runs.index(a)
    new = action(ac, s, w)    #saving the returned value from the function into a new list
    s = new[0]    #updating the value of new score
    w = new[1]    #updating the value of new wickets
    if(w==0):    #terminating the match if number of wickets left are 0
        break

print("Final score is " + str(s) + " and " + str(w) + " wickets left")    #printing the final result
```
